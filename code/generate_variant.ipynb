```python
def extract_characters(file_path):
    common_chars = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            parts = line.split()
            if len(parts) > 1:
                char = parts[1]

                if '\u4e00' <= char <= '\u9fff':
                    common_chars.append(char)
    return common_chars

file_path = 'result.txt'
common_chars = extract_characters(file_path)
print(f"common_chars = {common_chars}")
```

```python
import random

def get_dif_tone(char):
    pinyin = lazy_pinyin(char, style=Style.TONE3, neutral_tone_with_five=True)
    # print(pinyin)
    tone = pinyin[0][-1]
    py = pinyin[0][:-1]
    # print(py)

    dif_tone = []
    for c in common_chars:
        if c == char:
            continue
        c_pinyin = lazy_pinyin(c, style=Style.TONE3, neutral_tone_with_five=True)
        c_tone = c_pinyin[0][-1]
        c_py = c_pinyin[0][:-1]
        if c_py == py and c_tone != tone:
            dif_tone.append(c)
    # return dif_tone
    return random.choice(dif_tone) if dif_tone else char
```

```python
def get_same_char(char):
    pinyin = lazy_pinyin(char, style=Style.TONE3, neutral_tone_with_five=True)
    # print(pinyin)

    similar_chars = []
    for c in common_chars:
        if c == char:
            continue
        c_pinyin = lazy_pinyin(c, style=Style.TONE3, neutral_tone_with_five=True)
        if c_pinyin == pinyin:
            similar_chars.append(c)
    # return similar_chars
    return random.choice(similar_chars) if similar_chars else char
```

```python
def generate_variants(phrase):
    chars = list(phrase)

    # 1. 同音但替换一个字
    variant1 = chars.copy()
    pos = random.randint(0, len(chars) - 1)
    variant1[pos] = get_same_char(chars[pos])

    # 2. 同音且替换两个字
    variant2 = chars.copy()
    positions = random.sample(range(len(chars)), 2)
    for pos in positions:
        variant2[pos] = get_same_char(chars[pos])

    # 3. 一个字声调不同，且该字也被替换
    variant3 = chars.copy()
    pos = random.randint(0, len(chars) - 1)
    variant3[pos] = get_dif_tone(chars[pos])

    # 4. 两个字声调不同，且两个字也被替换
    variant4 = chars.copy()
    positions = random.sample(range(len(chars)), 2)
    for pos in positions:
        variant4[pos] = get_dif_tone(chars[pos])

    return [''.join(v) for v in [variant1, variant2, variant3, variant4]]
```

```python
import pandas as pd

input_file = "data_variant.csv"
output_file = "output.csv"
data = pd.read_csv(input_file)

first_column = data.iloc[:, 1]

variants_data = []

for cell in first_column:
    variants = generate_variants(cell)
    variants_data.append(variants)

variants_df = pd.DataFrame(variants_data, columns=["变体1", "变体2", "变体3", "变体4"])
variants_df.to_csv(output_file, index=False, encoding='utf-8')

print(f"处理完成，结果已保存到 {output_file}")
```
